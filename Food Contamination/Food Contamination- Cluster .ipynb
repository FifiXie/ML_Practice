## Converting Series to Str to take out stem words then cover back to ndArray

```python
type(raw_data['product_description'])
```

```python
s = pd.Series(raw_data['product_description'])
raw = s.to_string()
type(raw)
```

### Remove Special Characters and Numbers

```python
from nltk.tokenize import TreebankWordTokenizer
import re
corpus = raw
text=''
for word in re.findall(r'[a-zA-Z]+', corpus):
    text += word + ' '
    txt = text.strip()
```

### Remove any words that has 2 or less letters - (to avoid reoccuring word such as 'lb', 'oz', 'kg' )

```python
newDesc = ''
for word in txt.split():
    if len(word) > 2:
        newDesc += word + ' '
        newWords = newDesc.strip()
```

### Stemming words

```python
def stem(word):
    regexp = r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem, suffix = re.findall(regexp, word)[0]
    return stem

#raw = raw_data['product_description']
tokens = word_tokenize(newWords)
[stem(t) for t in tokens]
```

### Transfer Words into Matrix

```python
X_array = np.asarray(tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(strip_accents='unicode', stop_words='english', min_df=1)

#corpus = raw_data['product_description']
X_v = vectorizer.fit_transform(X_array)
X_v.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer(norm='l2', sublinear_tf=True)
X = transformer.fit_transform(X_v)
```

### Using KMeans to define clusters

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=38, init='k-means++', n_init=127, max_iter=900, tol=0.0001,  random_state=27, algorithm='full')
km.fit(X)
```
